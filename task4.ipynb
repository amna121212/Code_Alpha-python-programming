# Basic Chatbot

```python
def chatbot():
    print("🤖 Hello! I'm ChatBot. Type 'bye' to exit.\n")

    responses = {
        "hello": "Hi there! How can I help you?",
        "hi": "Hey! Nice to meet you ",
        "how are you": "I'm doing great, thanks for asking!",
        "what is your name": "I'm ChatBot, your simple Python assistant!",
        "what can you do": "I can answer basic questions and have a friendly chat!",
        "bye": "Goodbye! Have a great day! ",
        "thanks": "You're welcome! ",
        "help": "Try saying: hello, how are you, what is your name, bye",
    }

    while True:
        user_input = input("You: ").lower().strip()

        if not user_input:
            continue

        # Check for matching response
        matched = False
        for key in responses:
            if key in user_input:
                print(f"Bot: {responses[key]}\n")
                matched = True
                if key == "bye":
                    return
                break

        if not matched:
            print("Bot: Sorry, I didn't understand that. Type 'help' for options.\n")

chatbot()
```
