# Stock Portfolio Tracker

```python
import csv
import os

def stock_tracker():
    # Hardcoded stock prices
    stock_prices = {
        "AAPL": 180,
        "TSLA": 250,
        "GOOGL": 140,
        "MSFT": 320,
        "AMZN": 185
    }

    print("📈 Stock Portfolio Tracker")
    print("Available stocks:", ", ".join(stock_prices.keys()))
    
    portfolio = {}
    
    while True:
        stock = input("\nEnter stock symbol (or 'done' to finish): ").upper().strip()
        if stock == "DONE":
            break
        if stock not in stock_prices:
            print("❌ Stock not found. Try again.")
            continue
        try:
            qty = int(input(f"Enter quantity for {stock}: "))
            portfolio[stock] = portfolio.get(stock, 0) + qty
        except ValueError:
            print("❌ Invalid quantity.")

    if not portfolio:
        print("No stocks entered.")
        return

    print("\n📊 Portfolio Summary:")
    print(f"{'Stock':<10} {'Qty':<10} {'Price':<10} {'Value':<10}")
    print("-" * 40)
    
    total = 0
    for stock, qty in portfolio.items():
        value = qty * stock_prices[stock]
        total += value
        print(f"{stock:<10} {qty:<10} ${stock_prices[stock]:<9} ${value:<10}")
    
    print("-" * 40)
    print(f"{'TOTAL':<30} ${total}")

    # Save to CSV
    save = input("\nSave to file? (yes/no): ").lower()
    if save == "yes":
        filename = "portfolio.csv"
        with open(filename, "w", newline="") as f:
            writer = csv.writer(f)
            writer.writerow(["Stock", "Quantity", "Price", "Total Value"])
            for stock, qty in portfolio.items():
                writer.writerow([stock, qty, stock_prices[stock], qty * stock_prices[stock]])
            writer.writerow(["", "", "TOTAL", total])
        print(f"✅ Saved to {filename}")

stock_tracker()
```
