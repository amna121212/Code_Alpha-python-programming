# Task Automation — Move .jpg Files

```python
import os
import shutil

def move_jpg_files():
    print("📁 JPG File Mover Automation")

    source = input("Enter source folder path: ").strip()
    destination = input("Enter destination folder path: ").strip()

    if not os.path.exists(source):
        print("❌ Source folder does not exist.")
        return

    # Create destination if it doesn't exist
    os.makedirs(destination, exist_ok=True)

    jpg_files = [f for f in os.listdir(source) if f.lower().endswith(".jpg")]

    if not jpg_files:
        print("⚠️ No .jpg files found in source folder.")
        return

    moved = 0
    for file in jpg_files:
        src_path = os.path.join(source, file)
        dest_path = os.path.join(destination, file)
        shutil.move(src_path, dest_path)
        print(f"✅ Moved: {file}")
        moved += 1

    print(f"\n🎉 Done! {moved} file(s) moved to '{destination}'")

move_jpg_files()
```
